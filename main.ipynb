```python
from ultralytics import YOLO
import cv2
from google.colab import files
import numpy as np
import os
import requests

video_path = '/content/test1.mp4'
```

```python
model = YOLO("yolo11n.pt")
```

```python
models = {
    'YOLOv10n': YOLO('yolov10n.pt'),
    'YOLOv11n': YOLO("yolo11n.pt"),
    'YOLOv12n': YOLO('yolo12n.pt')
}

inference_speeds = {name: [] for name in models.keys()}
```

```python
def convert_avi_to_mp4(avi_path, mp4_path):
    cap = cv2.VideoCapture(avi_path)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(3))
    height = int(cap.get(4))
    out = cv2.VideoWriter(mp4_path, fourcc, fps, (width, height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        out.write(frame)

    cap.release()
    out.release()
    print(f'Converted to: {mp4_path}')
    return mp4_path
```

```python
for model_name, model in models.items():

    print(f'Running tracking with {model_name}...')
    results = model.track(source=video_path, show=False, save=True, persist=True, conf=0.25, iou=0.7, stream=True)

    # Stor inference speeds
    for r in results:
        inference_speeds[model_name].append(r.speed['inference'])

    # save it
    save_dir = r.save_dir
    avi_path = f'{save_dir}/test1.avi'
    mp4_path = f'{save_dir}/test1_{model_name.lower()}_tracked.mp4'

    convert_avi_to_mp4(avi_path, mp4_path)

    files.download(mp4_path)
    print(f'{model_name} results saved and downloaded as {mp4_path}')
```

```python
print('\nInference Speed Comparison (ms per frame):')
print('----------------------------------------')

for model_name in inference_speeds:
    avg_speed = np.mean(inference_speeds[model_name])
    print(f'{model_name}: {avg_speed:.2f}')
```
